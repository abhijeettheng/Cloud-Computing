# **Demo 03: Implementing return Statement**

**Objective:** To learn how to use the return statement in Python effectively

**Prerequisites:** None

**Tools required:** Python

### **Steps to be followed:**

1. Define a function with arguments
2. Use the return statement to return values when the function is called
3. Display the result


**Step 1:** Define a function with arguments

```python
def calculate_sum(a, b):
    """Function to add two numbers and return the result."""
    return ( a + b)
```

```python
result = calculate_sum(10, 20)
```

```python
print(result)
```

**Step 2:** Use the return statement to return values when the function is called

**Step 3:** Display the result

```python
print(f"The sum is: {result}")
```
