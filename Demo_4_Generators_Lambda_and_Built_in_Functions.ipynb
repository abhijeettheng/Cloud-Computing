**Step 1:** Define a generator function that yields values one by one

```python
def number_generator(n):
    """Generator function to yield numbers from 1 to n."""
    for i in range(1, n + 1):
        yield i  # Yielding values one by one
```

```python
# Using the generator function
gen = number_generator(5)
print("Generated numbers:")
for num in gen:
    print(num)
```

```python
# defination of generator
def myGenerator():
    n = 1
    print('First Iteration')
    yield 'Number 1 : n = ' + str(n)

    n += 2
    print('Second Iteration')
    yield 'Number 2 : n = ' + str(n)

    n += 3
    print('Third Iteration')
    yield 'Number 3 : n = ' + str(n)

    n = 1
    print('Forth Iteration')
    yield 'Number 3 : n = ' + str(n)
```

```python
gen=myGenerator()
```

```python
gen = myGenerator()
for value in gen :
    print(value, '\n')
```

**Step 2:** Introduce the lambda function and use it with built-in functions

```python
square = lambda x: x * x  # Lambda function to square a number
print(f" Square using lambda: {square(150)}")
```

```python
mul = lambda x,y,z: x * y *z
print(f" Multiplication of 3 numbers using lambda: {mul(41,4,2)}")
```

**Step 3:** Apply built-in functions like map() and filter() to process lists efficiently

```python
numbers = [1, 2, 3, 4, 5]
# Using map() to apply the lambda function
squared_numbers = list(map(lambda x: x ** 2, numbers))
print(f"Squared numbers using map(): {squared_numbers}")

```

```python
# Using filter() to get even numbers
even_numbers = list(filter(lambda x: x % 3, numbers))
print(f"Even numbers using filter(): {even_numbers}")
```

```python
from _functools import reduce
numbers = [2,4,6,8,10,12]
result = reduce(lambda x , y : x * y, numbers)
print('Total = ', result)
```
