# **Demo 07: Implementing File Handling in Python**

**Objective:** To understand file handling in Python, including reading, writing, appending, and processing files efficiently

**Prerequisites:** None

**Tools required:** Python

### **Steps to be followed:**

1. Read a file
2. Write to a file
3. Append to a file
4. Read after writing
5. Read line by line
6. Write strings and close the file

**Step 1:** Read a file

```python
# Step 1: Read a file (if it exists)
with open("example1.txt", "r") as file:
    print(file.read())  # Reading the entire file
```

**Step 2:** Write to a file

```python
# Step 2: Write to a file (overwrites existing content)
with open("example1.txt", "w") as file:
    file.write("overwriting.\n")
    file.write("Writing to a file in Python.\n")
```

**Step 3:** Append to a file

```python
# Step 3: Append to a file (adds new content without deleting existing content)
with open("example1.txt", "a") as file:
    file.write("Appending a new line.\n")
    file.write("Another appended line.\n")
```

**Step 4:** Read after writing

```python
# Step 4: Read after writing
with open("example.txt", "r") as file:
    print(file.read())
```

**Step 5:** Read line by line

```python
# Step 5: Read line by line
with open("example.txt", "r") as file:
    for line in file:
        print(line.strip("a"))
        print("Hello")
```

**Step 6:** Write strings and close the file

```python
# Step 6: Write strings and close the file manually
file = open("example.txt", "w")
file.write("Final content replacing everything.\n")
file.write("Closing the file explicitly.\n")
file.close()  # Manually closing the file
```

##### By following these steps, you have successfully learned how to handle files in Python, including reading, writing, appending, and processing files efficiently.
