# **Demo 05: Implementing a Class with Attributes And Methods**

**Objective:** To learn how to create a class and define methods and attributes in Python

**Prerequisites:** None

**Tools required:** Python

### **Steps to be followed:**

1. Create a class
2. Declare the desired attributes
3. Create a method that displays the information
4. Initiate the objects
5. Access class attributes and methods through objects


```python
# Step 1: Create a class
class Car:
    """A class to represent a car."""

    # Step 2: Declare the desired attributes
    def __init__(self, brand, model, year):
        self.brand = brand
        self.model = model
        self.year = year

    # Step 3: Create a method that displays the information
    def display_info(self):
        """Method to display car details."""
        print(f"Car: {self.year} {self.brand} {self.model}")
```

```python
# Step 4: Initiate the objects
car1 = Car("Toyota", "Corolla", 2022)
car2 = Car("Honda", "Civic", 2023)
```

```python
# Step 5: Access class attributes and methods through objects
car1.display_info()  # Output: Car: 2022 Toyota Corolla
car2.display_info()  # Output: Car: 2023 Honda Civic
```

By following these steps, you have successfully learned how to create a class, define methods, and set attributes in Python.
